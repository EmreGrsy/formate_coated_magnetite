```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd

# Define a Gaussian function
def gauss(x, mu, sigma, A):
    return A * np.exp(-(x - mu)**2 / (2 * sigma**2))

# Define a model for multiple Gaussians
def multi_gauss(x, *params):
    num_gauss = len(params) // 3
    y = np.zeros_like(x)
    for i in range(num_gauss):
        mu = params[3*i]
        sigma = params[3*i + 1]
        A = params[3*i + 2]
        y += gauss(x, mu, sigma, A)
    return y

# Function to fit the histograms with multiple Gaussians
def fit_histogram(data, ax, num_gaussians):
    # Generate the histogram
    y, x = np.histogram(data, bins=30)
    x = (x[1:] + x[:-1]) / 2  # Midpoints of bins
    
    # Initial guess for the parameters (mu, sigma, amplitude for each Gaussian)
    params_init = []
    for i in range(num_gaussians):
        params_init += [np.mean(data) + i*0.1, np.std(data), max(y)/(i+1)]
    
    # Fit the model
    params, cov = curve_fit(multi_gauss, x, y, p0=params_init)
    
    # Plot the data
    ax.hist(data, bins=30, color="blue", edgecolor="black", alpha=0.5)
    
    # Plot the fit
    x_fit = np.linspace(x.min(), x.max(), 500)
    y_fit = multi_gauss(x_fit, *params)
    ax.plot(x_fit, y_fit, color="red", lw=2, label='Fit')
    
    # Plot individual Gaussians
    for i in range(num_gaussians):
        y_individual = gauss(x_fit, params[3*i], params[3*i + 1], params[3*i + 2])
        ax.plot(x_fit, y_individual, lw=1, ls='--')
    
    ax.legend()

# Example of applying this function to your datasets
fig, ax = plt.subplots(6, 2, figsize=(12, 20), sharex=True, sharey=True)

datasets_tet = [d1, d2, d3, d4, d5, d6]
datasets_int = [d7, d8, d9, d10, d11, d12]

for i in range(6):
    fit_histogram(datasets_tet[i].distance1_Fe2Fe3, ax[i][0], num_gaussians=3)
    fit_histogram(datasets_int[i].distance1_Fe2Fe3, ax[i][1], num_gaussians=3)

ax[5][0].set_xlabel("Tet")
ax[5][1].set_xlabel("Int")

plt.tight_layout()
plt.show()
```
