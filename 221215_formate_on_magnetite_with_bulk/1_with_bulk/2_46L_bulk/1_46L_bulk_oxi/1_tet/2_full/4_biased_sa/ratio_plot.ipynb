```python
font = {"size"   : 20}
plt.rc("font", **font)
plt.rcParams['ytick.right']
df = pd.read_csv("ratio.data",
                 header = None, 
                 delimiter=" ")


fig,axis = plt.subplots(figsize=(2.5, 4),facecolor="w")
axis.invert_yaxis()

axis.set_xticks([0, 0.5, 1.0])
#axis.set_xlim([-0.15, 1.15])

#axis.set_yticks([0, 5, 10])

axis.plot(df[1]/36, df[0], linewidth=4, color="blue")

axis.set_xlabel(r"$n_{\mathrm{{Fe}_{oct}^{3+}}}/\mathrm{n_{{Fe}_{oct}}}$", labelpad=20, fontsize=24)

axis.set_ylabel(r"$\mathrm{{Fe}_{oct}}$ Layer", labelpad=15, fontsize=24)

axis.set_title(r"33L, Full-Tet" "\n" r"Both Surface", pad=20, fontsize=24)
```
